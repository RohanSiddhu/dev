```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")
data1 = pd.read_csv("data1.csv")
data2 = pd.read_csv("data2.csv")

data_input = data['input_size'].to_numpy()
data_divide_and_conquer = data['divide_and_conquer'].to_numpy()
data_brute_force = data['brute_force'].to_numpy()

data1_input = data1['input_size'].to_numpy()
data1_divide_and_conquer = data1['divide_and_conquer'].to_numpy()
data1_brute_force = data1['brute_force'].to_numpy()

data2_input = data2['input_size'].to_numpy()
data2_divide_and_conquer = data2['divide_and_conquer'].to_numpy()
data2_brute_force = data2['brute_force'].to_numpy()

x = np.linspace(1, 1000, 1000)
y = x * np.log2(x)
y2 = x ** 2
```

```python
fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y2)
ax2.plot(data_input, data_brute_force)

fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y)
ax2.plot(data_input, data_divide_and_conquer)

fig3, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y2)
ax1.plot(x, y)
ax2.plot(data_input, data_brute_force)
ax2.plot(data_input, data_divide_and_conquer)

plt.show()
```

```python
fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y2)
ax2.plot(data1_input, data1_brute_force)

fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y)
ax2.plot(data1_input, data1_divide_and_conquer)

fig3, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(x, y2)
ax1.plot(x, y)
ax2.plot(data1_input, data1_brute_force)
ax2.plot(data1_input, data1_divide_and_conquer)

plt.show()
```
